```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(-10, 10, 0.1)
y = np.arange(0, 200, 2)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = X ** 2 + Y
```

```python
# %%capture
fig = plt.figure(figsize = [19.2, 14.4])
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
```

```python
surf1 = ax1.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm)
surf2 = ax2.plot_surface(X, Y, Z + np.random.normal(0, 10, Z.shape), cmap=plt.cm.coolwarm)
```

```python
ax1.set_zlabel('z', fontsize=15)
ax1.set_xlabel('x', fontsize=15)
ax1.set_ylabel('y', fontsize=15)
cb1 = fig.colorbar(surf1, ax=ax1, aspect=20, shrink=0.3)
```
