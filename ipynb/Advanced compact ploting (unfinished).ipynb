```python
data = pdr.DataReader('AAPL', 'yahoo')
```

```python
df = data[data.columns[-2:]]
```

```python
df
```

```python
fig = plt.figure(figsize=[9.6, 9.6])
```

```python
ax0 = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4, fig=fig)
ax1 = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4, fig=fig, sharex=ax0)
ax1.set_facecolor('#fce4d6')
l0, = ax0.plot(df['Adj Close'])
plt.setp(ax0.get_xticklabels(), visible=False)
fig.subplots_adjust(hspace=0)
# or plt.subplots_adjust if fig declaration is in the same cell

ax1t = ax0.twinx()
ax1t.fill_between(df.index, df['Volume'], alpha=0.5)
ax1t.set_ylim(0, df['Volume'].max() * 3)
ax1t.set_yticks([df['Volume'].max()])
ax1t.grid(False)
```
